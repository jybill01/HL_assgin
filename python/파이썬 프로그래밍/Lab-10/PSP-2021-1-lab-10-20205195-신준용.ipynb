#### 문제 1.

다음은 $\sin(2\pi t)$의 그래프이다.
아래 그래프를 그리는 Python 프로그램을 작성하라.

그림의 크기는 가로 8 인치, 세로 3인치이고, 선과 마커의 색은 r=0, g=0.3, b=0.6이다.

<img src='fig-sin(2 pi t).png' />

```python
t = np.linspace(0, 1, 31)
x = np.sin(2*np.pi*t)
plt.figure(figsize = (8, 3))
plt.plot(t, x, 's--', color= (0, 0.3, 0.6))
plt.xticks([0, 0.5, 1.0])
plt.xlabel('$t$')
plt.xlim(0, 1)
plt.title('graph of $sin(2\pi t)$')
plt.show()
```

#### 문제 2.

다음은 $\cos(t)$의 그래프이다.
아래 그래프를 그리는 Python 프로그램을 작성하라.

그림의 크기는 가로 8 인치, 세로 3인치이고, 선과 마커의 색은 'cyan'이다.

<img src='fig-cos(t).png' />

```python
t = np.linspace(0, 2*np.pi, 21)
x = np.cos(t)
plt.figure(figsize = (8, 3))
plt.plot(t, x, 'cd-.', color= 'cyan')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], [0, r'$\frac{\pi}{2}$', '$\pi$', r'$\frac{3\pi}{2}$', '$2\pi$'])
plt.xlabel('$t$')
plt.xlim(0, 2*np.pi)
plt.title('graph of $cos(t)$')
plt.show()
```

#### 참고: 3차원 그래프 그리기

```python
%matplotlib inline
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib
import matplotlib.pyplot as plt
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def f(x, y, mx, my, sx, sy):
    return 1/(sx*sx*2*np.pi)*np.exp(-(x-mx)**2/(2*sx**2) - (y-my)**2/(2*sy**2))*np.abs(np.sin(y))

x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)
mx = 0
my = 0
sx = 0.5
sy = 1.5

X,Y = np.meshgrid(x, y)
Z = f(X, Y, mx, my, sx, sy)

fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1, 2, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=True)
cb = fig.colorbar(p, shrink=0.5)
plt.show()                    
```
