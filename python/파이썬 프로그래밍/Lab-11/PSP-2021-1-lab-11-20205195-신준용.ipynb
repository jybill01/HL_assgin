##### 9.5
scores에  각 학생의 점수 합계를 기록한 새로운 column '합계'를 추가하라.

````python
scores['합계'] = scores[['시험1', '시험2', '시험3', '시험4', '시험5']].sum(axis = 1)
scores
````

##### 4.6
'합계'의 내림차순으로 scores를 정렬하라.

````python
scores.sort_values(by = '합계', ascending = False, inplace = True)
scores
````

##### 9.7
 각 소속별 학생들의 총점의 평균 점수를 막대 그래프로 나타내라.

````python
total_scores = scores[['소속', '합계']]
total_mean_per_dept = total_scores.groupby('소속').mean()
total_mean_per_dept.sort_values(by = '합계', ascending = False, inplace = True)
total_mean_per_dept
````

````python
total_mean = []
for i in range(total_mean_per_dept.values.size):
    total_mean.append(round(total_mean_per_dept.values[i][0]))

plt.bar(range(total_mean_per_dept.values.size), total_mean)
plt.xticks(range(total_mean_per_dept.values.size), total_mean_per_dept.index, rotation = 90)
plt.ylabel('평균') 
plt.show()
````

##### 9.8
 학생 수가 가장 많은 소속의 학생들의 정보를 화면에 표시하라.

````python
scores.groupby('소속').size().argmax()
````

````python
amx_dept = scores.groupby('소속').size().idxmax()
scores[scores['소속'] == amx_dept]
````

##### 9.9
이 수업을 듣는 학생들의 시험 성적 사이의 상관계수(correlation coefficient)를 구하라.

````python
plt.scatter(scores['시험1'], scores['시험2'])
plt.show()
````

````python
scores['시험1'].corr(scores['시험3'])
````

````python
a = ['시험1', '시험2', '시험3', '시험4', '시험5']

for i in a:
    for j in a:
        plt.scatter(scores[j], scores[i])
        plt.xlabel(j)
        plt.ylabel(i)
        plt.title(f"{i}와 {j}의 상관계수 : {scores[i].corr(scores[j])}")
        plt.show()
````
