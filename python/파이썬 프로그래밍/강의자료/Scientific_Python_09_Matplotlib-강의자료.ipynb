#### Matplotlib
- Python과 NumPy를 위한 plotting library 
- 범용 GUI toolkits를 사용하는 applications에 내장할 수 있는 객체 지향 API
- Matlab와 매우 닮은 절차적 Pylab 인터페이스

#### Matplotlib.pyplot
- matplotlib가 MATLAB처럼 작동하도록하는 명령 스타일 함수들의 모음
- 한 문장의 명령을 추가할 때마다 figure에 변화 발행

### Line Plot

##### matplot.pyplot.plot()

2차원 평면 그래프에서 다음 점들을 선분으로 연결하여 그래프를 그리는 함수.

$$P_0(x_0, y_0), P_1(x_1, y_1), \cdots , P_{N-1}(x_{N-1}, y_{N-1})$$ 

```
plot([x0, x1, ..., x_{N-1}], [y0, y1, ..., y_{N-1}] )
```

#### 예제  1.

다음 네 점을 연결하는 그래프를 그려라.

$$(0, 1), (1, -1), (2, 1), (3, -1)$$

#### matplot.pyplot.figure()를 이용한 figure의 크기 지정
```
plt.figure(figsize=(width, height))
```

- figsize : (float, float), optional, default: None
- width, height in inches. 
  - 기본값: rcParams["figure.figsize"] = [6.4, 4.8]


````python
plt.figure(figsize=(8,3))
x = np.arange(0, 1, 0.01)
y = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)
plt.plot(x, y, linestyle='-', color='cyan', label='$\sin(2\pi x)$')
plt.plot(x, y2, ls='-.', color='m', label='$\cos(2\pi x)$')
plt.xlim(0,1)
plt.ylim(-1,1)
plt.legend(loc=9)
plt.title("Plots of sine and cosine")

ax = plt.gca() # get the current Axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
#ax.yaxis.set_ticks_position('left')
#ax.spines['left'].set_position(('data',0))




plt.show()
````

#### plt.subplot(): 현재의 figure에 새로운 axes를 추가

````python
x = np.linspace(0, np.pi*2, 201)
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = np.sin(3*x)
y4 = np.sin(4*x)
y5 = np.sin(5*x)
y6 = np.sin(6*x)
````

````python
plt.figure(figsize=(10,6))
plt.subplot(2, 3, 1)
plt.plot(x, y1)
plt.subplot(2, 3, 2)
plt.plot(x, y2)
plt.subplot(2, 3, 3)
plt.plot(x, y3)
plt.subplot(234)
plt.plot(x, y4)
plt.subplot(235)
plt.plot(x, y5)
plt.subplot(236)
plt.plot(x, y6)
plt.show()
````

````python
plt.figure(figsize=(10,6))
plt.subplot(2, 3, 1)
plt.plot(x, y1)
plt.title("$\sin(t)$")
plt.subplot(2, 3, 2)
plt.plot(x, y2)
plt.title("$\sin(2t)$")
plt.subplot(2, 3, 3)
plt.plot(x, y3)
plt.title("$\sin(3t)$")
plt.subplot(234)
plt.plot(x, y4)
plt.title("$\sin(4t)$")
plt.subplot(235)
plt.title("$\sin(5t)$")
plt.plot(x, y5)
plt.subplot(236)
plt.plot(x, y6)
plt.title("$\sin(6t)$")
plt.tight_layout()
plt.show()
````

#### 3차원 그래프 그리기

````python
%matplotlib inline
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib
import matplotlib.pyplot as plt
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def f(x, y, mx, my, sx, sy):
    return 1/(sx*sx*2*np.pi)*np.exp(-(x-mx)**2/(2*sx**2) - (y-my)**2/(2*sy**2))*np.abs(np.sin(y))

x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)
mx = 0
my = 0
sx = 0.5
sy = 1.5

X,Y = np.meshgrid(x, y)
Z = f(X, Y, mx, my, sx, sy)

fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1, 2, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=True)
cb = fig.colorbar(p, shrink=0.5)
plt.show()                    
````
