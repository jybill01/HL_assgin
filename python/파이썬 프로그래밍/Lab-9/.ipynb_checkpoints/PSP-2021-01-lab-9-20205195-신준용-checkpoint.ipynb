#### 문제 1. 3원색 원 그리기

##### 1.1) 다음 cell에서 img1, img2, img3의 shape을 설명하고, 그 안에 어떤 데이터가 있는지 설명하라.

````python
cols = 201
rows = 201
img1 = np.full((rows,cols,3), 0, np.uint8)
img2 = np.full((rows,cols,3), 0, np.uint8)
img3 = np.full((rows,cols,3), 0, np.uint8)
````

**답**:(201, 201, 3)3차원 배열에 0으로 이루어져 있다.

##### 2.2) 다음 cell에서 idx_x, idx_y의 shape을 설명하고, 그 안에 어떤 데이터가 있는지 설명하라.

````python
cr = np.array([65, 65])
cg = np.array([65, 135])
cb = np.array([126, 100])
r = 50

idx_x = np.arange(cols).reshape(-1,1)
idx_y = np.arange(rows).reshape(1,-1)
print(idx_x.shape, idx_y.shape)
````

**답**:idx_x는 (201, 1) = 세로, idx_y는 (1, 201) = 가로, 로 배열되고 0 ~ 200까지 데이터가 차례대로 들어있다

##### 2.3) 다음 cell에서 img1 안에 어떤 데이터가 있는지 설명하라.

````python
img1[(idx_x-cr[0])**2 + (idx_y-cr[1])**2 < r**2] = np.array([255, 0, 0])
````

````python
plt.imshow(img1)
plt.show()
````

**답**:원의 중심을 (65,65)로 잡고 idx_x와 idx_y에서 각각 65를 빼고 제곱하면 중심으로 부터 거리를 구할 수 있는데 이를 통해서 중심으로 부터 반지름인 50의 제곱보다 작은 데이터를 (0, 0, 0)에서 (255, 0, 0)으로 바꾸었고 나머지는 그대로 (0, 0, 0)이다

##### 2.4) 다음 cell에서 img2와 img3 안에 각각 어떤 데이터가 있는지 설명하라.

````python
img2 = np.full_like(img1, (0, 0, 0))
img3 = np.full_like(img1, (0, 0, 0))
````

````python
img2[(idx_x-cg[0])**2 + (idx_y-cg[1])**2 < r**2] = np.array([0, 255, 0])
````

````python
plt.imshow(img2)
plt.show()
````

**답**:img3는 (0, 0, 0)으로만 이루어져 있고 img2는 (65, 135)를 중심으로 거리의 제곱이 반지름 제곱보다 작은 데이터는 (0, 255, 0)으로 이루어져 있고 나머지는 (0, 0, 0)이다.

##### 2.5) 다음 cell에서 노란색 부분에 있는 한 픽셀에는 어떤 어떤 데이터가 있는지 설명하라.

````python
img12 = img1 + img2
plt.imshow(img12)
plt.show()
````

**답**:(255, 0, 0) 과 (0, 255, 0)이 합쳐져서 (255, 255, 0)로 이루어져있다.

##### 2.6) 다음 cell에서 원들이 겹치는 부분의 색이 각각 yellow, magenta, cyan, white 색이 되는 이유를 설명하라.

````python
img3[(idx_x-cb[0])**2 + (idx_y-cb[1])**2 < r**2] = np.array([0, 0, 255])
````

````python
plt.imshow(img3)
plt.show()
````

````python
img123 = img1 + img2 + img3
plt.imshow(img123)
plt.show()
````

**답**:(255, 0, 0) 과 (0, 255, 0)과 (0, 0, 255)가 합쳐져서 중앙에 데이터가 (255, 255, 255)가 되기 떄문에 하얀색이 된다.

##### 2.7) img123을 png 파일 포맷으로 저장한 후에 다시 읽어서 화면에 표시하라.

````python
imageio.imsave("tri-color-circles.png", img123)
````

````python
plt.imshow(f)
plt.show()
````

#### 문제 3. NumPy ndarray를 이용하여 다음 영상을 만들어서 화면에 표시하라.

<img src='bw_check.png' width="35%"/>

````python
bw_chk = np.full((10, 10), 0, np.uint8)
idx_x = np.arange(10).reshape(-1,1)
idx_y = np.arange(10).reshape(1,-1)
bw_chk[(idx_x + idx_y) % 2 != 0] = np.array([255])
plt.imshow(bw_chk, cmap='gray')
plt.show()
````

#### 문제 4. NumPy ndarray를 이용하여 다음 영상을 만들어서 화면에 표시하라.

<img src='color_check.png' width="40%"/>

````python
a = np.full((6, 15, 3), (255,255,255) , np.uint8)
col = [[255,0,0],[0,255,0],[0,0,255],[0,255,255],[255,0,255],[255,255,0]]
for i in range(len(a)):
    for j in range(len(a[i])):
        if((j % 3) == (i % 3)):
            a[i][j] = col[i]
plt.imshow(a)
plt.show()
````

#### 문제 6. Histogram 구하기

##### 6.1) 아래 셀에 있는 shape이 (5,5)인 배열로 표현되는 영상 s_img의 각 숫자의 개수를 프로그램을 사용하지 말고 손으로 직접 계산하라.


````python
np.random.seed(153)

s_img = np.random.randint(1, 5, (5,5))
s_img
````

**답**

- 1: 7
- 2: 5
- 3: 9
- 4: 4

##### 6.2) 문제 6.1)에서 구한 데이터를 이용해서 다음 히스토그램의 그래프를 그리는 Python 프로그램을 작성하라.



<img src="histogram-4x4.png" width="40%"/>


````python
n = np.arange(1, 5)
h = np.array([7, 5, 9, 4])
plt.bar(n, h, width = 0.98)
plt.xticks(n)
plt.show()
````

##### 6.3) s_img 배열의 히스토그램을 구하는 Python 프로그램을 작성하라.

````python
n = np.arange(1, 5)
s_img_hist = np.zeros_like(n)
for p in n:
    s_img_hist[p - 1] = np.sum(s_img == p)
s_img_hist
````

````python
np.sum(s_img == 1)
````

````python
plt.bar(n, s_img_hist)
plt.xticks(n)
plt.show()
````

####  문제 8.

"nature-3082832_640.jpg" 이미지를 흑백 이미지로 변환하여 화면에 표시하라.

````python
img_gray2 = (img[:,:,0].astype(np.float) + img[:,:,1].astype(np.float) + img[:,:,2].astype(np.float)) / 3
img_gray2 = img_gray2.astype(np.uint8)
plt.imshow(img_gray2, cmap='gray')
plt.show()
````

####  문제 9.

"nature-3082832_640.jpg" 이미지의 흑백 이미지의 히스토그램을 그려라.

####  문제 11.

"nature-3082832_640.jpg" 이미지의 흑백 이미지를 $f(x,y)$라고 할 때, 다음 변환을 통해서 새로운 이미지 $g(x,y)$를 만들어 화면에 표시하라.

$$g(x,y) = \begin{cases}
0, & f(x,y) < 52 \\
255/150(f(x,y) - 52), & 52 \le f(x, y) \le 202 \\
0, &f(x, y) > 202
\end{cases}
$$

````python
n = img_gray
gf = 255 / 150 * (n - 52)
gf[n < 52] = 0
gf[n > 202] = 0

plt.imshow(gf, cmap = 'gray')
plt.show()
````
